```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns',None)
```

```python
firefighting = pd.read_csv('fire_firefighting.csv')
weather = pd.read_csv('weather_final.csv')
```

```python
def analyze_dataframe(df):

    print("데이터프레임 정보 : ")
    display(df.info(verbose=True, show_counts=True))
    print("\n")

    print("데이터프레임 값 확인 : ")
    display(df.head(5).T)
    print("\n")

    print("데이터프레임 통계 :")
    display(df.describe().T)
    print("\n")

    print("데이터프레임 누락 갯수 :")
    display(df.isnull().sum())
    print("\n")

    print("데이터프레임 중복 행 갯수 :")
    display(df.duplicated().sum())
    print("\n")

    print("데이터프레임 고유값 갯수 : ")
    display(df.nunique())
    print("\n")

    print("데이터프레임 형태 :")
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
analyze_dataframe(weather)
```

```python
drop_col = ['Unnamed: 0', 'Unnamed: 0.1','fire_weather.stn','fire_weather.district_2', '지역명']
weather.drop(drop_col, axis=1, inplace=True)
```

```python
weather.columns
```

```python
weather = weather.rename(columns={'district' :'지역명'})
```

```python
firefighting.drop(columns='Unnamed: 0', inplace=True)
firefighting['일시'] = pd.to_datetime(firefighting['fire_firefighting.tm']).dt.date
firefighting['지역명'] = firefighting['fire_firefighting.district_1'] + ' ' + firefighting['fire_firefighting.district_2']
```

```python
firefighting.loc[(firefighting['fire_firefighting.fire_type_1'] == '임야') & (firefighting['fire_firefighting.fire_type_2'].isna()), 'fire_firefighting.fire_type_2'] = '임야'
firefighting.loc[(firefighting['fire_firefighting.fire_type_1'] == '기타(쓰레기 화재등)') & (firefighting['fire_firefighting.fire_type_2'].isna()), 'fire_firefighting.fire_type_2'] = '기타(쓰레기 화재등)'
```

```python
firefighting.rename(columns = {'fire_firefighting.num' :'연번', 
                      'fire_firefighting.year' : '연도', 
                      'fire_firefighting.tm' : '화재발생시간', 
                      'fire_firefighting.district_1' : '시도명', 
                      "fire_firefighting.district_2" :'시군구명', 
                      "fire_firefighting.fire_type_1" :'화재유형 대분류', 
                      'fire_firefighting.fire_type_2' : '화재유형 소분류', 
                      'fire_firefighting.ignition_factor_category_1' : '발화요인대분류', 
                      'fire_firefighting.ignition_factor_category_2' : '발화요인소분류', 
                      'fire_firefighting.casualties' : '인명피해소계', 
                      'fire_firefighting.dead' : '사망', 
                      'fire_firefighting.injury' : '부상', 
                      'fire_firefighting.property_damage' : '재산피해소계', 
                      'fire_firefighting.location_category_1' : '장소대분류', 
                      'fire_firefighting.location_category_2' : '장소중분류', 
                      'fire_firefighting.location_category_3' : '장소소분류'}, inplace = True)
```

```python
weather['일시'] = pd.to_datetime(weather['일시']).dt.date
```

```python
merged_data = pd.merge(weather, firefighting, on=['일시', '지역명'], how='left')
column_order = ['일시', '지역명'] + [col for col in merged_data.columns if col not in ['일시', '지역명']]
merged_data = merged_data[column_order]
```

```python
merged_data
```

```python
merged_data['화재 여부'] = np.where(merged_data['화재유형 대분류'].isna(), 0, 1)
```

```python
merged_data[merged_data['화재 여부'] == 0]
```

```python
merged_data.fillna('-', inplace=True)
```

```python
analyze_dataframe(merged_data)
```

```python
spearman_corr = merged_data[['최고기온(℃)', '최저기온(℃)', '일교차', '강수량(mm)', '최대풍속(m/s)',
       '최대순간풍속(m/s)', '평균풍속(m/s)', '평균습도(%rh)', '최저습도(%rh)', '화재 여부']].corr(method='spearman')
print(spearman_corr)
```

```python
import statsmodels.api as sm

X = merged_data[['최고기온(℃)', '최저기온(℃)', '일교차', '강수량(mm)', '최대풍속(m/s)',
       '최대순간풍속(m/s)', '평균풍속(m/s)', '평균습도(%rh)', '최저습도(%rh)']]
X = sm.add_constant(X)
y = merged_data['화재 여부']

model = sm.Logit(y, X).fit()
print(model.summary())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)

predictions = clf.predict(X_test)

print(accuracy_score(y_test, predictions))
print(classification_report(y_test, predictions))
```
